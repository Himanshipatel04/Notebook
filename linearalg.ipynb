```python
def get_age_group(age):
    bins = [0, 18, 30, 50, float('inf')]
    labels = ['minor', 'adult', 'senior', 'aged']
    return pd.cut([age], bins=bins, labels=labels, right=False)[0]

# Apply the function to create the 'age_group' column
dff['age_group'] = dff['age'].apply(get_age_group)
```

```python
dff
```

```python
sns.countplot(x="age_group",data=dff,hue="platform",palette='Set2')
```
