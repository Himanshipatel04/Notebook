```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv(r"C:\Users\himan\Downloads\Desktop\tested.csv")
```

```python
df
```

```python
df.drop(columns = ["Name","Cabin"], inplace=True , axis = 1 )
```

```python
df["Age"]=df['Age'].fillna(df['Age'].mean())
```

```python
df
```

```python
df['Agei']=df['Age'].astype("int")
df.drop(columns="Age",axis=1,inplace=True)
```

```python
df
```

```python
new_column_name = 'Age'
df.columns = [new_column_name if col == 'Agei' else col for col in df.columns]
```

```python
df
```

```python
df["Sex"]=df["Sex"].astype('category')
```

```python
df.info()
```

```python
df['sex']=df['Sex'].cat.codes
```

```python
enc=OneHotEncoder()
```

```python
data=pd.DataFrame(enc.fit_transform(df[['sex']]).toarray())
```

```python
data
```

```python
newdf=df.join(data)
```

```python
newdf
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(newdf,hue="sex",height=10)
plt.show()
```
